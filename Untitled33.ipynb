```python
!pip install faker
import pandas as pd
import numpy as np
from faker import Faker
import random
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
# Initialize Faker
fake = Faker()
np.random.seed(42)
# Generate Dataset
num_rows = 5000
categories = ['Furniture', 'Office Supplies', 'Technology']
regions = ['East', 'West', 'Central', 'South']
data = {
    'Order_ID': [fake.unique.uuid4()[:8] for _ in range(num_rows)],
    'Customer_Name': [fake.name() for _ in range(num_rows)],
    'Category': [random.choice(categories) for _ in range(num_rows)],
    'Region': [random.choice(regions) for _ in range(num_rows)],
    'Quantity': np.random.randint(1, 10, num_rows),
    'Price': np.round(np.random.uniform(50, 2000, num_rows), 2),
    'Discount': np.round(np.random.uniform(0, 0.5, num_rows), 2),
    'Order_Date': [fake.date_between(start_date='-2y', end_date='today') for _ in range(num_rows)]
}
df = pd.DataFrame(data)
# Calculate Sales
df['Sales'] = (df['Quantity'] * df['Price']) * (1 - df['Discount'])
# Introduce Missing Values (for transformation example)
for col in ['Customer_Name', 'Discount']:
    df.loc[df.sample(frac=0.02).index, col] = np.nan
# Introduce Duplicates
df = pd.concat([df, df.sample(50)])
print("Initial Dataset Shape:", df.shape)
missing_before = df.isnull().sum().sum()
df['Customer_Name'] = df['Customer_Name'].fillna('Unknown')
df['Discount'] = df['Discount'].fillna(df['Discount'].mean())
missing_after = df.isnull().sum().sum()
print(f"Missing values handled: {missing_before - missing_after}")
duplicates_before = df.duplicated().sum()
df.drop_duplicates(inplace=True)
duplicates_after = df.duplicated().sum()
print(f"Duplicates removed: {duplicates_before - duplicates_after}")
rows_before = df.shape[0]
Q1 = df['Price'].quantile(0.25)
Q3 = df['Price'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Price'] >= (Q1 - 1.5 * IQR)) & (df['Price'] <= (Q3 + 1.5 * IQR))]
rows_after = df.shape[0]
print(f"Outliers removed: {rows_before - rows_after}")
df['Total_Sales'] = df['Sales'].round(2)
print("Feature engineered: Added 'Total_Sales' column")
encoder = LabelEncoder()
df['Category_Encoded'] = encoder.fit_transform(df['Category'])
df['Region_Encoded'] = encoder.fit_transform(df['Region'])
print("Categorical encoding applied: 'Category' and 'Region'")
scaler = MinMaxScaler()
df[['Price_Scaled', 'Quantity_Scaled']] = scaler.fit_transform(df[['Price', 'Quantity']])
print("Scaling applied to 'Price' and 'Quantity'")
print("\nFinal Dataset Shape:", df.shape)
print(df.head())
df.to_csv("sales_dataset_transformed.csv", index=False)
print("\nDataset saved as 'sales_dataset_transformed.csv'")
```
